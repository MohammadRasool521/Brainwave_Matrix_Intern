# SUPERSTORES SALES DASHBOARD

```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading csv file with pandas
df = pd.read_csv("C:\\Power BI\\Reports\\SuperStore_Sales_Dataset.csv")
df.head()
```

```python
df.describe()
```

```python
# Histogram
plt.hist(df['Region'], bins=20, color='Blue')
# adding title
plt.title('Sales by Region')
# adding X, Y- axis Labels
plt.xlabel('Region')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
#Bar Chart between Payment Mode and Sales
plt.bar(df['Payment Mode'],df['Sales'])
#adding title
plt.title('Sales by Payment Mode')
#adding X, Y-axis Labels
plt.xlabel('Payment Mode')
plt.ylabel('Sales')
plt.show()
```

```python
sns.boxplot(x='Region',y='Profit',notch=True,data=df,showfliers=False,color='red')
#adding Title
plt.title('Profits by Region')
# adding X, Y- axis Labels
plt.xlabel('Region')
plt.ylabel('Profits')
plt.show()
```

```python
# Scatter Plot between Region and Sales
plt.scatter(df['Region'],df['Sales'])
# adding title
plt.title('Sales_by_Region')
# adding X, Y- axis Labels
plt.xlabel('Region')
plt.ylabel('Sales')
# adding more customization
plt.colorbar()
plt.show()
```
